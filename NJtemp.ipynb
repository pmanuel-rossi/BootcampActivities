```python
import matplotlib.pyplot as plt

avgTemp = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]

months = ['Jan', 'Feb', 'Mar', 'Apr','May', 'Jun', 'Jul', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']

plt.plot(months, avgTemp)


plt.show()
```

```python
avgtempCelsius = [(f - 32) * 0.56 for f in avgTemp]

plt.plot(months, avgtempCelsius)
```

```python
plt.plot(months, avgTemp, label = 'Fahrenheit', marker= '+')
plt.plot(months, avgtempCelsius, label ='Celsius' , marker= 's')

plt.legend()

plt.xlabel("Months")
plt.ylabel("Temperature")

plt.show()
```

```python
import matplotlib.pyplot as plt

runtime = list(range(0,130,10))

dangerDrop = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]

railGun = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]

plt.xlim(min(runtime), max(runtime))

plt.plot(runtime, dangerDrop, label= 'Danger Drop')

plt.plot(runtime, railGun, label= 'Rail Gun')

plt.title('Coaster Speed Over Time')
plt.xlabel('Coaster Runtime')
plt.ylabel('Speed (MPH)')
plt.grid()
plt.legend()
```
